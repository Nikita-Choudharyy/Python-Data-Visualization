```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
tips.info()
```

# 📈 Line Plot

A Line Plot is used to visualize how a numerical value changes over an ordered variable, such as time or any sequential data. It helps identify trends, patterns, increases, decreases, and fluctuations in the data.

### 📌 Common Use Cases

- 📅 Time series analysis
- 📈 Tracking growth or decline over time
- 💹 Monitoring trends and performance
- 🔄 Visualizing continuous data

### 🌍 Real-World Example

A company wants to analyze its monthly sales over the past year. A Line Plot makes it easy to identify whether sales are increasing, decreasing, or remaining consistent over time.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a line plot showing the relationship between party size and total bill
sns.lineplot(
    data=tips,
    x="size",
    y="total_bill"
)

# Add a descriptive title
plt.title("Relationship Between Party Size and Total Bill")

# Label the x-axis
plt.xlabel("Party Size")

# Label the y-axis
plt.ylabel("Total Bill ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The x-axis represents the party size, while the y-axis represents the average total bill.
- The total bill generally increases as the party size increases.
- Smaller parties (1–2 people) tend to have lower average bills.
- Larger parties (5–6 people) usually spend more, resulting in higher total bills.
- The shaded region represents the confidence interval, showing the variability in the average total bill for each party size.
- The wider shaded area for larger party sizes suggests greater variation in customer spending.

# ⚫ Scatter Plot

A Scatter Plot is used to visualize the relationship between two numerical variables. Each point on the graph represents a single observation, making it easy to identify patterns and relationships within the data.

### 📌 Common Use Cases

- 📈 Identifying correlation between variables
- 🎯 Detecting clusters or groups
- 🚨 Finding outliers
- 📊 Discovering trends and patterns

### 🌍 Real-World Example

A teacher wants to analyze whether students who study more hours score higher marks in an exam. A Scatter Plot helps visualize the relationship between study hours and exam scores and identify any positive or negative correlation.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a scatter plot showing the relationship between total bill and tip amount
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

# Add a descriptive title
plt.title("Relationship Between Total Bill and Tip")

# Label the x-axis
plt.xlabel("Total Bill ($)")

# Label the y-axis
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- Each point represents a single customer.
- There is a positive relationship between the total bill and the tip amount.
- As the total bill increases, the tip amount generally tends to increase.
- Most customers have total bills between **$10 and $25**, with tips ranging from **$2 to $4**.
- A few customers have much higher bills and tips, indicating possible outliers.
- The spread of points increases for higher total bills, suggesting greater variation in tip amounts.

# 📊 Bar Plot

A Bar Plot is used to compare numerical values across different categories. In Seaborn, it displays the average (mean) value for each category by default, making it easy to compare groups.

### 📌 Common Use Cases

- 📊 Comparing values across categories
- 📈 Comparing average measurements
- 🛍️ Analyzing sales across different products
- 👥 Comparing customer behavior across groups

### 🌍 Real-World Example

A restaurant wants to compare the average total bill during lunch and dinner. A Bar Plot makes it easy to see which meal time has the higher average bill.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a bar plot showing the average total bill for each day
# By default, Seaborn calculates the mean for each category.
sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Average Total Bill by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Average Total Bill ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The bar plot compares the **average total bill** for each day of the week.
- Sunday has the highest average total bill, followed closely by Saturday.
- Thursday and Friday have lower average total bills compared to the weekend.
- The results suggest that customers tend to spend more on weekends than on weekdays.
- The vertical error bars represent the confidence interval around the mean, showing the variability of the estimated average for each day.

# 🔢 Count Plot

A Count Plot displays the number of observations in each category. Unlike a Bar Plot, it does **not** calculate an average or any other statistic—it simply counts how many times each category appears in the dataset.

### 📌 Common Use Cases

- 📊 Counting observations in each category
- 👥 Analyzing the distribution of categorical data
- 🛍️ Comparing customer categories
- 📈 Understanding the frequency of different groups

### 🌍 Real-World Example

A restaurant wants to know on which day it receives the most customer visits. A Count Plot shows the number of bills recorded for each day, making it easy to compare customer traffic.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a count plot showing the number of customers for each day
sns.countplot(
    data=tips,
    x="day"
)

# Add a descriptive title
plt.title("Number of Customers by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Number of Customers")

# Add a light grid for better readability
plt.grid(axis="y", alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- The count plot shows the number of customers (observations) recorded for each day.
- Saturday has the highest number of customer visits in the dataset.
- Sunday has the second-highest number of customer visits.
- Thursday has a moderate number of customer visits.
- Friday has the fewest customer visits among all the days.
- Overall, customer traffic is higher on weekends than on weekdays.

> **💡 Note:** A Count Plot displays the **frequency (count)** of each category. It does **not** calculate averages, totals, or other statistical measures.

# 📊 Histogram

A Histogram is used to visualize the distribution of a numerical variable by dividing the data into intervals called **bins**. It shows how frequently values fall within each range, making it easier to understand the overall distribution of the data.

### 📌 Common Use Cases

- 📈 Understanding the distribution of numerical data
- 🔢 Analyzing the frequency of values
- 📉 Identifying skewness and symmetry
- 🚨 Detecting possible outliers
- 📊 Examining the spread of the data

### 🌍 Real-World Example

A retail company wants to analyze the distribution of customer purchase amounts. A Histogram helps identify whether most customers spend a small, moderate, or large amount and whether there are any unusually high purchases.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a histogram showing the distribution of total bill amounts
sns.histplot(
    data=tips,
    x="total_bill",
    bins=20
)

# Add a descriptive title
plt.title("Distribution of Total Bill")

# Label the x-axis
plt.xlabel("Total Bill ($)")

# Label the y-axis
plt.ylabel("Frequency")

# Add a light grid for better readability
plt.grid(axis="y", alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- The histogram shows the distribution of customer total bill amounts.
- Most total bills fall between **$10 and $20**, indicating that this is the most common spending range.
- The frequency of bills gradually decreases as the bill amount increases.
- Only a few customers have total bills above **$40**, making high-value bills relatively uncommon.
- The distribution is **right-skewed (positively skewed)**, with a longer tail extending toward higher bill amounts.
- There are no obvious gaps in the distribution, suggesting that the bill amounts are spread continuously across the range.

> **💡 Note:** Each bar represents the number of observations (frequency) within a specific range of values called a **bin**. Changing the number of bins changes the level of detail shown in the histogram.

# 📈 KDE Plot

A KDE (Kernel Density Estimation) Plot is used to visualize the distribution of a continuous numerical variable. It creates a smooth curve that estimates how the data is distributed, making it easier to identify patterns without relying on histogram bins.

### 📌 Common Use Cases

- 📊 Understanding the distribution of numerical data
- 📈 Identifying peaks in the data
- 📉 Detecting skewness and symmetry
- 🔍 Comparing multiple distributions
- 📊 Visualizing the overall shape of the data

### 🌍 Real-World Example

A retail company wants to analyze customer purchase amounts. A KDE Plot provides a smooth view of the spending distribution, helping identify where most customers' purchases are concentrated.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a KDE plot showing the distribution of total bill amounts
sns.kdeplot(
    data=tips,
    x="total_bill",
    fill=True
)

# Add a descriptive title
plt.title("Distribution of Total Bill (KDE Plot)")

# Label the x-axis
plt.xlabel("Total Bill ($)")

# Label the y-axis
plt.ylabel("Density")

# Add a light grid for better readability
plt.grid(alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- The KDE plot shows a smooth estimate of the distribution of total bill amounts.
- The highest density occurs between **$10 and $20**, indicating that most customer bills fall within this range.
- The density gradually decreases as the total bill amount increases.
- The distribution is **right-skewed (positively skewed)**, with a longer tail extending toward higher bill amounts.
- Bills above **$40** are relatively uncommon, as indicated by the low density in that range.
- The smooth curve makes it easier to identify the overall distribution pattern compared to a histogram.

> **💡 Note:** A KDE Plot estimates the distribution of continuous data using a smooth curve. Unlike a Histogram, it does **not** divide the data into bins, making it useful for understanding the overall shape of the distribution.

# 📦 Box Plot

A Box Plot summarizes the distribution of a numerical variable using the **five-number summary**. It helps quickly understand the spread of the data, identify the center of the distribution, and detect potential outliers.

### 📌 What a Box Plot Shows

- 📍 Minimum value (excluding outliers)
- 📦 First Quartile (Q1)
- ➖ Median (Q2)
- 📦 Third Quartile (Q3)
- 📍 Maximum value (excluding outliers)
- 🚨 Potential outliers

### 📌 Common Use Cases

- 📊 Understanding the spread of numerical data
- 📈 Comparing distributions across different groups
- 🎯 Identifying the median of the data
- 🚨 Detecting outliers
- 📉 Measuring data variability

### 🌍 Real-World Example

A retail company wants to compare customer purchase amounts across different days of the week. A Box Plot quickly shows the median spending, spread of the data, and any unusually high or low purchases.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a box plot showing the distribution of total bill by day
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Distribution of Total Bill by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Total Bill ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The box plot compares the distribution of total bill amounts across different days of the week.
- The median total bill is highest on **Sunday**, followed closely by **Saturday**.
- Saturday and Sunday have a wider spread (larger interquartile range), indicating greater variation in customer spending.
- Thursday and Friday have relatively lower median bills and a smaller spread compared to the weekend.
- Several points lie above the upper whiskers, indicating **potential outliers** (customers with unusually high total bills).
- Overall, customers tend to spend more and show greater variability in spending on weekends than on weekdays.

> **💡 Note:** In a Box Plot, the **box** represents the middle 50% of the data (Q1 to Q3), the **line inside the box** represents the median, the **whiskers** show the typical range of the data, and the **individual points** beyond the whiskers indicate potential outliers.

# 🎻 Violin Plot

A Violin Plot combines the features of a **Box Plot** and a **KDE Plot**. It summarizes the distribution of a numerical variable while also showing the probability density of the data, making it easier to understand both the spread and the shape of the distribution.

### 📌 What a Violin Plot Shows

- 📦 Distribution of the data
- 📈 Density of values
- ➖ Median
- 📊 Spread and variability
- 🔍 Shape of the distribution (e.g., skewness or multiple peaks)

### 📌 Common Use Cases

- 📊 Comparing distributions across different categories
- 📈 Understanding the shape of the data
- 🎯 Identifying where values are concentrated
- 📉 Detecting skewness and variability
- 🔍 Comparing multiple groups in a single visualization

### 🌍 Real-World Example

A restaurant wants to compare customer spending across different days of the week. A Violin Plot not only shows the median and spread of the total bill but also reveals where customer spending is most concentrated on each day.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a violin plot showing the distribution of total bill by day
sns.violinplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Distribution of Total Bill by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Total Bill ($)")

# Add a light grid for better readability
plt.grid(axis="y", alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- The violin plot compares the distribution of total bill amounts across different days of the week.
- The wider sections of each violin indicate where customer bills are more concentrated.
- Saturday and Sunday show a wider distribution, suggesting greater variability in customer spending.
- Thursday and Friday have relatively narrower distributions, indicating that most customer bills fall within a smaller range.
- The median total bill appears to be higher on weekends than on weekdays.
- Overall, customer spending is more varied on weekends, while weekday spending is relatively more consistent.

> **💡 Note:** Unlike a Box Plot, a Violin Plot also shows the **density** of the data. Wider parts of the violin represent a higher concentration of observations, while narrower parts indicate fewer observations.

# 📍 Strip Plot

A Strip Plot displays every individual observation in the dataset. Each data point is plotted separately, making it easy to see the distribution, spread, and concentration of values within different categories.

### 📌 Common Use Cases

- 📊 Visualizing individual observations
- 📈 Comparing distributions across categories
- 🔍 Identifying clusters and gaps
- 🚨 Detecting potential outliers
- 📉 Understanding the spread of the data

### 🌍 Real-World Example

A restaurant wants to examine the total bill for every customer across different days of the week. A Strip Plot displays each customer's bill as an individual point, making it easy to observe the spread and distribution of spending.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a strip plot showing individual total bill observations for each day
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Distribution of Total Bill by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Total Bill ($)")

# Add a light grid for better readability
plt.grid(axis="y", alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- Each dot represents the total bill for a single customer.
- The plot shows the distribution of individual bill amounts across different days of the week.
- Most customer bills are concentrated between **$10 and $25** across all days.
- Saturday and Sunday have a larger number of observations and a wider spread of bill amounts compared to Thursday and Friday.
- A few customers have unusually high total bills (above **$40**), which appear as individual points away from the main cluster.
- The Strip Plot allows us to observe every individual data point, making it easier to identify the spread, clusters, and potential outliers within each day.

> **💡 Note:** In a Strip Plot, each point represents one observation. The slight horizontal spread of the points (called **jitter**) helps reduce overlap, making individual observations easier to see.

# 🐝 Swarm Plot

A Swarm Plot is similar to a Strip Plot but automatically adjusts the position of points to prevent overlap. This makes every individual observation easier to distinguish, especially when many data points have similar values.

### 📌 Common Use Cases

- 📊 Visualizing individual observations without overlap
- 📈 Comparing distributions across categories
- 🔍 Identifying clusters and gaps
- 🚨 Detecting potential outliers
- 📉 Understanding the spread of the data

### 🌍 Real-World Example

A restaurant wants to compare the total bill for every customer across different days of the week. A Swarm Plot displays each customer's bill as a separate point while preventing overlapping, making the distribution easier to interpret.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a swarm plot showing individual total bill observations for each day
sns.swarmplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Distribution of Total Bill by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Total Bill ($)")

# Add a light grid for better readability
plt.grid(axis="y", alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- Each point represents the total bill for a single customer.
- The Swarm Plot displays all individual observations without overlapping points, making the distribution easier to interpret.
- Saturday and Sunday have more customer observations and a wider range of total bill amounts than Thursday and Friday.
- Most customer bills are concentrated between **$10 and $25** across all days.
- A few customers have unusually high total bills (above **$40**), which appear as isolated points.
- Compared to the Strip Plot, the Swarm Plot provides a clearer view of individual observations by preventing point overlap.

> **💡 Note:** Unlike a Strip Plot, a Swarm Plot automatically arranges points so they do not overlap. This makes it easier to identify clusters, gaps, and potential outliers, especially in datasets with many observations.

# 📚 Boxen Plot

A Boxen Plot (Letter Value Plot) is an extension of the **Box Plot** that provides a more detailed view of the distribution of numerical data. It displays additional quantiles, making it especially useful for exploring large datasets and understanding the distribution of values in greater detail.

### 📌 Common Use Cases

- 📊 Analyzing the distribution of large datasets
- 📈 Comparing distributions across different categories
- 🔍 Understanding the spread of the data
- 🚨 Detecting potential outliers
- 📉 Identifying patterns in the tails of the distribution

### 🌍 Real-World Example

A retail company wants to compare customer purchase amounts across different store locations. A Boxen Plot provides a more detailed view of the spending distribution than a standard Box Plot, making it easier to identify variations and unusual spending patterns.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a boxen plot showing the distribution of total bill by day
sns.boxenplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Distribution of Total Bill by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Total Bill ($)")

# Add a light grid for better readability
plt.grid(axis="y", alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- The Boxen Plot compares the distribution of total bill amounts across different days of the week.
- The median total bill is higher on **Saturday** and **Sunday** than on **Thursday** and **Friday**.
- Weekend bill amounts show a wider distribution, indicating greater variation in customer spending.
- The multiple nested boxes reveal additional details about the distribution, especially in the lower and upper portions of the data.
- Several unusually high bill amounts are visible on **Saturday** and **Sunday**, suggesting more high-value customer purchases during the weekend.
- Compared to a standard Box Plot, the Boxen Plot provides a more detailed view of how the data is distributed across different quantiles.

> **💡 Note:** A Boxen Plot extends a Box Plot by displaying multiple quantiles instead of only the five-number summary. This provides a more detailed understanding of the distribution, particularly for large datasets.

# 📍 Point Plot

A Point Plot displays the estimated value of a numerical variable for each category. By default, Seaborn plots the **mean** of each category along with **confidence intervals**, making it useful for comparing values and identifying trends across categories.

### 📌 Common Use Cases

- 📊 Comparing average values across categories
- 📈 Identifying trends between categories
- 📉 Visualizing confidence intervals
- 🔍 Comparing multiple groups
- 📋 Summarizing categorical data

### 🌍 Real-World Example

A restaurant wants to compare the **average total bill** across different days of the week. A Point Plot clearly shows the average bill for each day along with confidence intervals, making it easy to compare customer spending patterns.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Draw a point plot showing the average total bill for each day
sns.pointplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Average Total Bill by Day")

# Label the x-axis
plt.xlabel("Day")

# Label the y-axis
plt.ylabel("Average Total Bill ($)")

# Add a light grid for better readability
plt.grid(axis="y", alpha=0.3)

# Display the plot
plt.show()
```

### 📌 Observation

- The Point Plot shows the **average total bill** for each day of the week.
- The average total bill is highest on **Sunday**, followed closely by **Saturday**.
- **Friday** has the lowest average total bill among all the days.
- There is an increasing trend in the average total bill from **Friday** to **Sunday**.
- The vertical error bars represent the **confidence intervals** around the estimated mean for each day.
- The confidence intervals overlap slightly, suggesting that the differences between some days may not be statistically significant.

> **💡 Note:** A Point Plot displays the estimated value (typically the **mean**) for each category along with **confidence intervals**. It is useful for comparing averages and identifying trends across categories.

# 🔗 Pair Plot

A Pair Plot visualizes the pairwise relationships between numerical variables in a dataset. It automatically creates scatter plots for every pair of numerical features and displays the distribution of each feature along the diagonal.

It is one of the most commonly used visualizations during **Exploratory Data Analysis (EDA)** because it provides a quick overview of relationships between multiple variables.

### 📌 Common Use Cases

- 📊 Exploring relationships between numerical variables
- 📈 Identifying correlations and trends
- 🔍 Detecting clusters and patterns
- 🚨 Identifying potential outliers
- 📉 Understanding the distribution of each feature

### 🌍 Real-World Example

A data analyst is exploring a customer dataset containing **Age, Annual Income, Spending Score, and Purchase Amount**. A Pair Plot quickly reveals how these variables relate to one another, helping identify correlations, clusters, and unusual observations before building a machine learning model.

```python
# Create a pair plot to visualize relationships between numerical features
sns.pairplot(
    data=tips,
    vars=["total_bill", "tip", "size"],
    hue="sex"
)

# Display the plot
plt.show()
```

### 📌 Observation

- The Pair Plot displays the relationships between **total bill**, **tip**, and **party size**, with observations colored by **sex**.
- A positive relationship is visible between **total bill** and **tip**, indicating that customers with higher bills generally leave higher tips.
- The relationship between **party size** and **total bill** is moderately positive, with larger groups tending to have higher total bills.
- The diagonal KDE plots show that most customers have **moderate total bills**, **tips around $2–4**, and **party sizes of 2 to 3 people**.
- The distributions for **Male** and **Female** customers overlap considerably, suggesting similar overall spending and tipping patterns.
- A few observations with unusually high total bills and tips are visible, indicating potential high-spending customers.

> **💡 Note:** A Pair Plot provides both the **distribution of individual variables** (diagonal plots) and the **relationships between variables** (off-diagonal scatter plots). The `hue` parameter helps compare different categories by displaying them in different colors.

# 🔥 Heatmap

A Heatmap is a visualization that represents values using colors. Different colors indicate different magnitudes, making it easy to identify patterns, relationships, and trends within the data.

Heatmaps are one of the most commonly used visualizations in **Exploratory Data Analysis (EDA)**.

### 📌 Common Use Cases

- 📊 Visualizing correlation matrices
- 📉 Detecting missing values
- 🤖 Evaluating confusion matrices
- 📋 Displaying pivot tables
- 🔍 Identifying patterns and relationships in data

### 🌍 Real-World Example

A data analyst wants to understand how different numerical features in a customer dataset are related. A correlation heatmap quickly highlights strong positive, negative, and weak relationships between variables, helping guide feature selection before building a machine learning model.

```python
# Calculate the correlation matrix for numerical columns
correlation = tips.corr(numeric_only=True)

# Create a figure with a custom size
plt.figure(figsize=(8, 6))

# Draw a heatmap to visualize the correlation matrix
sns.heatmap(
    correlation,
    annot=True,        # Display correlation values inside each cell
    cmap="coolwarm",   # Use a color gradient from cool to warm
    fmt=".2f"          # Display values with 2 decimal places
)

# Add a descriptive title
plt.title("Correlation Heatmap")

# Display the plot
plt.show()
```

### 📌 Observation

- The heatmap displays the correlation between the numerical variables: **Total Bill**, **Tip**, and **Party Size**.
- **Total Bill** and **Tip** have a **strong positive correlation (0.68)**, indicating that customers with higher bills generally leave higher tips.
- **Total Bill** and **Party Size** have a **moderate positive correlation (0.60)**, suggesting that larger groups tend to generate higher total bills.
- **Tip** and **Party Size** show a **moderate positive correlation (0.49)**, indicating that larger parties tend to leave slightly higher tips.
- The diagonal values are **1.00** because each variable is perfectly correlated with itself.
- No negative correlations are present, meaning all numerical variables tend to increase together to some extent.

> **💡 Note:** Correlation values range from **-1** to **1**.
>
> - **1.00** → Perfect positive correlation
> - **0.70 to 0.99** → Strong positive correlation
> - **0.30 to 0.69** → Moderate positive correlation
> - **0.00** → No linear correlation
> - **-0.30 to -0.69** → Moderate negative correlation
> - **-0.70 to -1.00** → Strong negative correlation
>
> Correlation measures the strength and direction of a **linear relationship** between two numerical variables. A high correlation **does not imply causation**.

# 📊 Choosing the Right Plot

Different plots are designed for different types of data and analysis. Choosing the appropriate visualization helps communicate insights more effectively.

| Plot Type | Best Used For |
|-----------|---------------|
| 📈 Line Plot | Visualizing trends over time or ordered data |
| ⚫ Scatter Plot | Exploring relationships between two numerical variables |
| 📊 Bar Plot | Comparing average or aggregated values across categories |
| 🔢 Count Plot | Displaying the frequency of each category |
| 📉 Histogram | Understanding the distribution of a numerical variable |
| 📈 KDE Plot | Visualizing the smooth probability density of a numerical variable |
| 📦 Box Plot | Summarizing distributions and detecting potential outliers |
| 🎻 Violin Plot | Comparing distributions while showing data density |
| 📍 Strip Plot | Displaying every individual observation |
| 🐝 Swarm Plot | Displaying individual observations without overlapping points |
| 📚 Boxen Plot | Exploring detailed distributions in large datasets |
| 📍 Point Plot | Comparing category averages and identifying trends |
| 🔗 Pair Plot | Exploring relationships among multiple numerical variables |
| 🔥 Heatmap | Visualizing correlations, matrices, and patterns using colors |

# 💡 Best Practices

When creating Seaborn visualizations:

- Choose the most appropriate plot for your data and analysis goal.
- Keep visualizations simple, clean, and easy to interpret.
- Add clear titles, axis labels, and legends where necessary.
- Use colors consistently and avoid overly bright or distracting color palettes.
- Avoid overcrowding charts with excessive information or unnecessary elements.
- Combine Seaborn with Matplotlib for advanced customization.
- Validate insights using multiple visualizations instead of relying on a single plot.
- Ensure charts are readable by adjusting figure size, font size, and spacing.
- Always interpret the visualization in the context of the data rather than relying only on visual patterns.
- Remember that **correlation does not imply causation** when interpreting relationships between variables.

# ❌ Common Mistakes

Avoid these common mistakes when creating data visualizations:

- Choosing a plot that is not suitable for the type of data or analysis.
- Ignoring outliers without investigating their cause.
- Overusing colors, which can make charts difficult to interpret.
- Forgetting to add clear titles, axis labels, and legends.
- Drawing conclusions from a single visualization instead of validating insights with multiple plots.
- Skipping data cleaning and quality checks before creating visualizations.
- Misinterpreting correlation as causation.
- Creating cluttered charts with unnecessary elements or excessive information.
