# 🎨 Seaborn Themes

A **theme** controls the overall appearance of Seaborn visualizations, including the background, grid lines, axes, and overall style.

Choosing the right theme improves readability and makes charts more suitable for analysis, presentations, and reports.

### 📌 Available Themes

- **darkgrid** – Dark background with grid lines, ideal for general data analysis.
- **whitegrid** – White background with grid lines, commonly used for statistical visualizations.
- **dark** – Dark background without grid lines, suitable for minimalist designs.
- **white** – Clean white background without grid lines, useful for presentations and publications.
- **ticks** – White background with axis ticks, commonly used in academic and publication-quality figures.

### 🌍 Real-World Example

A business analyst may use the **whitegrid** theme for dashboards because the grid lines make values easier to compare, while a researcher preparing a paper may prefer the **ticks** theme for a cleaner and publication-ready appearance.

```python
# Set the Seaborn theme to "darkgrid"
sns.set_theme(style="darkgrid")

# Create a scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

# Add a descriptive title
plt.title("Scatter Plot with Darkgrid Theme")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **darkgrid** theme adds subtle grid lines that make it easier to estimate values along both axes.
- The scatter plot shows a **positive relationship** between **Total Bill** and **Tip**, with customers who spend more generally leaving higher tips.
- Most observations are concentrated between **$10–25** for the total bill and **$2–4** for the tip.
- A few customers have unusually high total bills and tips, which appear as potential outliers.
- The grid lines improve readability without making the visualization appear cluttered, making the chart well-suited for exploratory data analysis (EDA).

> **💡 Note:** The **darkgrid** theme is Seaborn's default style. It adds light grid lines that make it easier to compare values and estimate positions, making it an excellent choice for exploratory data analysis.

```python
# Set the Seaborn theme to "whitegrid"
sns.set_theme(style="whitegrid")

# Create a scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

# Add a descriptive title
plt.title("Scatter Plot with Whitegrid Theme")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **whitegrid** theme applies a clean white background with subtle grid lines, making the chart easy to read.
- The grid lines help estimate values more accurately without distracting from the data.
- A **positive relationship** is visible between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- Most observations are concentrated between **$10–25** for the total bill and **$2–4** for the tip.
- A few customers have unusually high total bills and tips, appearing as potential high-value observations.
- The clean white background gives the visualization a professional appearance, making it suitable for reports, dashboards, and presentations.

> **💡 Note:** The **whitegrid** theme combines a clean white background with subtle grid lines, making it one of the most popular Seaborn themes for statistical analysis and business reporting.

```python
# Set the Seaborn theme to "white"
sns.set_theme(style="white")

# Create a scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

# Add a descriptive title
plt.title("Scatter Plot with White Theme")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **white** theme provides a clean white background without grid lines, giving the visualization a simple and uncluttered appearance.
- The scatter plot shows a **positive relationship** between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- Most observations are concentrated between **$10–25** for the total bill and **$2–4** for the tip.
- A few observations with unusually high total bills and tips are visible, representing potential high-spending customers.
- The absence of grid lines places the primary focus on the data points, resulting in a clean and minimalist visualization.

> **💡 Note:** The **white** theme removes grid lines and background elements, making it ideal for clean, minimalist visualizations. While it improves visual simplicity, it may be slightly harder to estimate exact values compared to themes with grid lines.

# 📐 Plot Context

A **plot context** controls the scaling of visual elements in a Seaborn visualization. It adjusts the size of text, lines, markers, and other plot components without changing the underlying data.

Different contexts are useful depending on where the visualization will be displayed, such as notebooks, presentations, posters, or research papers.

### 📌 What Changes with Context?

- 🔠 Font size
- 📏 Line width
- 🔵 Marker size
- 📊 Overall scaling of plot elements

### 📌 Available Contexts

- **paper** – Small elements, ideal for research papers and printed documents.
- **notebook** – Default context, optimized for Jupyter notebooks and general analysis.
- **talk** – Larger elements, suitable for presentations and classroom demonstrations.
- **poster** – Largest elements, designed for posters and large displays.

### 🌍 Real-World Example

A data scientist may use the **notebook** context while exploring data, switch to **talk** when presenting results in a meeting, and use **paper** when preparing figures for a research publication.

```python
# Set the Seaborn context to "paper"
sns.set_context("paper")

# Create a scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

# Add a descriptive title
plt.title("Scatter Plot with Paper Context")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **paper** context displays smaller text, markers, and plot elements, creating a compact and space-efficient visualization.
- The scatter plot continues to show a **positive relationship** between **Total Bill** and **Tip**, indicating that customers with higher bills generally leave higher tips.
- Most observations remain concentrated between **$10–25** for the total bill and **$2–4** for the tip.
- The reduced element sizes make the chart suitable for printed documents while maintaining good readability.
- The context changes only the **appearance** of the visualization; the data, trend, and interpretation remain unchanged.

> **💡 Note:** The **paper** context is optimized for research papers, reports, and documentation. It reduces the size of plot elements so that figures fit well in printed or digital documents without sacrificing clarity.

```python
# Set the Seaborn context to "talk"
sns.set_context("talk")

# Create a scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

# Add a descriptive title
plt.title("Scatter Plot with Talk Context")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **talk** context increases the size of the title, axis labels, tick labels, and markers, making the visualization easier to read.
- The scatter plot continues to show a **positive relationship** between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- Most observations remain concentrated between **$10–25** for the total bill and **$2–4** for the tip.
- Larger markers make individual data points more visible, especially when presenting the chart to an audience.
- The overall visualization is more suitable for presentations, lectures, and meetings where readability from a distance is important.
- The context changes only the **appearance** of the plot elements; the underlying data and insights remain the same.

> **💡 Note:** The **talk** context is designed for presentations and live demonstrations. It automatically enlarges text, markers, and other plot elements so that charts remain clear and readable when displayed on larger screens.

# 🌈 Color Palettes

A **color palette** is a collection of colors used to represent data in a visualization. Choosing an appropriate palette improves readability, highlights important information, and makes charts more visually appealing.

Seaborn provides several built-in color palettes designed for different visualization needs.

### 📌 Built-in Color Palettes

- **deep** – Default palette with balanced, professional colors for general-purpose visualizations.
- **muted** – Softer colors that reduce visual intensity and are suitable for subtle presentations.
- **pastel** – Light, soft colors that create a clean and friendly appearance.
- **bright** – Highly saturated colors that emphasize categories and attract attention.
- **dark** – Rich, darker colors that provide strong contrast.
- **colorblind** – A colorblind-friendly palette designed to improve accessibility for a wider audience.

### 🌍 Real-World Example

A business analyst may use the **deep** palette for dashboards, **pastel** for educational content, **bright** to highlight important categories in presentations, and **colorblind** when creating visualizations for a diverse audience to ensure the charts remain accessible.

## 🔍 Palette Comparison

| Palette | Appearance | Best Used For |
|---------|------------|---------------|
| **deep** | Balanced and professional | General-purpose visualizations |
| **muted** | Soft and subtle | Reports and dashboards |
| **pastel** | Light and gentle | Educational content and presentations |
| **bright** | Vibrant and eye-catching | Highlighting important categories |
| **dark** | Rich and high contrast | Dark-themed visualizations |
| **colorblind** | Accessible color scheme | Inclusive and accessible visualizations |

```python
# Set the Seaborn color palette to "deep"
sns.set_palette("deep")

# Create a bar plot
sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Bar Plot with Deep Palette")

# Label the axes
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **deep** palette uses balanced and moderately saturated colors, giving the chart a clean and professional appearance.
- The bar plot compares the **average total bill** across different days of the week.
- **Sunday** has the highest average total bill, followed closely by **Saturday**.
- **Thursday** and **Friday** have comparatively lower average total bills.
- The error bars indicate the uncertainty (confidence interval) around the estimated average for each day.
- The **deep** palette improves visual appeal while maintaining excellent readability without using overly bright colors.

> **💡 Note:** The **deep** palette is Seaborn's default color palette. It provides balanced colors that are easy on the eyes, making it an excellent choice for general-purpose visualizations, dashboards, and Exploratory Data Analysis (EDA).

```python
# Set the Seaborn color palette to "pastel"
sns.set_palette("pastel")

# Create a bar plot
sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

# Add a descriptive title
plt.title("Bar Plot with Pastel Palette")

# Label the axes
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **pastel** palette applies soft, light colors that create a clean and visually pleasing appearance.
- The bar plot compares the **average total bill** across different days of the week.
- **Sunday** has the highest average total bill, followed closely by **Saturday**.
- **Thursday** and **Friday** have comparatively lower average total bills.
- The lighter color tones make the visualization appear calm and less visually intense while preserving readability.
- The error bars represent the confidence interval around the estimated average total bill for each day.

> **💡 Note:** The **pastel** palette uses light, low-saturation colors that create a soft and friendly appearance. It is ideal for educational content, reports, and presentations where a gentle visual style is preferred.

# 📏 Figure Size

Although Seaborn is used to create visualizations, the **figure size** is controlled using **Matplotlib** through the `plt.figure()` function.

Adjusting the figure size helps improve readability by providing enough space for titles, labels, legends, and data points.

### 📌 Why Change the Figure Size?

- 📊 Improve the readability of charts.
- 📝 Prevent overlapping labels and titles.
- 📈 Display large datasets more clearly.
- 📄 Create figures suitable for reports and publications.
- 🎥 Optimize visualizations for presentations and dashboards.

### 🌍 Real-World Example

A data analyst may use a **smaller figure size** for a report with limited space and a **larger figure size** for presentations or dashboards where the audience needs to view the chart from a distance.

```python
# Create a figure with a custom size (width=10 inches, height=6 inches)
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

# Add a descriptive title
plt.title("Scatter Plot with Customized Figure Size")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The figure size has been increased to **10 × 6 inches**, providing more space for the visualization.
- The larger figure makes the **title, axis labels, and data points** easier to read.
- The scatter plot shows a **positive relationship** between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- Most observations are concentrated between **$10–25** for the total bill and **$2–4** for the tip.
- A few observations with high total bills and tips appear in the upper-right corner, representing potential high-spending customers.
- The increased figure size improves readability and reduces visual crowding without changing the underlying data or insights.

> **💡 Note:** Changing the figure size affects only the **appearance** of the visualization. It provides additional space for titles, labels, legends, and data points, making charts easier to read without altering the data.

# 🔲 Grid Customization

Grid lines improve the readability of visualizations by making it easier to compare values and estimate the position of data points.

Although Seaborn provides built-in themes with grid styles, **Matplotlib** offers additional control over the appearance of grid lines, such as their visibility, style, color, and transparency.

### 📌 Why Customize Grid Lines?

- 📊 Improve value comparison across the chart.
- 👀 Enhance readability without cluttering the visualization.
- 📈 Make trends and patterns easier to identify.
- 🎨 Match the style of reports, dashboards, or presentations.
- 📄 Create professional and publication-ready figures.

### 🌍 Real-World Example

A financial analyst may use subtle grid lines in a sales report to help readers compare monthly revenue values. In contrast, a presentation slide may use lighter or fewer grid lines to keep the focus on the main insights.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Create a line plot
sns.lineplot(
    data=tips,
    x="size",
    y="total_bill"
)

# Customize the grid
plt.grid(
    True,
    linestyle="--",
    alpha=0.5
)

# Add a descriptive title
plt.title("Line Plot with Customized Grid")

# Label the axes
plt.xlabel("Party Size")
plt.ylabel("Total Bill ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The customized **dashed grid lines** make it easier to compare total bill values across different party sizes.
- The line plot shows an **overall increasing trend**, indicating that the average total bill generally increases as the party size becomes larger.
- The steepest increase is observed between **party sizes 2 and 4**, where the average total bill rises noticeably.
- There is only a small increase between **party sizes 4 and 5**, followed by another increase for **party size 6**.
- The shaded region around the line represents the **confidence interval**, showing the uncertainty around the estimated average total bill.
- The light, semi-transparent grid improves readability without distracting from the data.

> **💡 Note:** The `plt.grid()` function only changes the appearance of the grid lines—it does **not** affect the plotted data or the statistical information shown in the chart. Using a light grid style improves readability while keeping the focus on the data.

# 🏷️ Legend Customization

A **legend** explains the meaning of colors, markers, or line styles used in a visualization. It helps readers quickly identify different groups represented in the chart.

Seaborn automatically creates legends when using semantic mappings such as **`hue`**, **`style`**, or **`size`**. Matplotlib can then be used to customize the legend's title, position, size, and appearance.

### 📌 Why Customize Legends?

- 🏷️ Clearly identify different categories or groups.
- 👀 Improve the readability of complex visualizations.
- 📊 Make charts easier to interpret.
- 🎨 Position the legend to avoid covering important data.
- 📄 Create clean, presentation-ready visualizations.

### 🌍 Real-World Example

A sales dashboard may use different colors to represent customer segments such as **Regular**, **Premium**, and **VIP**. A well-positioned legend allows readers to quickly identify each group without confusion.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Create a scatter plot with a legend based on gender
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex"
)

# Add a descriptive title
plt.title("Scatter Plot with Customized Legend")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Customize the legend
plt.legend(
    title="Gender",
    loc="upper left"
)

# Display the plot
plt.show()
```

### 📌 Observation

- The scatter plot uses **different colors** to distinguish between **Male** and **Female** customers based on the `sex` column.
- The customized legend is titled **"Gender"**, making it clear what each color represents.
- The legend is placed in the **upper-left corner**, where it is easy to read and does not obstruct the majority of the data points.
- Both male and female customers show a **positive relationship** between **Total Bill** and **Tip**, indicating that tips generally increase as the total bill increases.
- Male customers appear more frequently in the dataset and are spread across a wider range of total bill and tip values, while female customers are comparatively fewer.
- The customized legend improves the readability and interpretation of the visualization without affecting the underlying data.

> **💡 Note:** The `hue` parameter automatically creates a legend for categorical variables. Customizing the legend title and position makes the chart easier to understand and helps prevent the legend from overlapping important data points.

# 🎨 Using `hue`

The **`hue`** parameter adds color to a visualization based on a **categorical variable**, making it easy to compare different groups within the same chart.

Instead of creating separate plots for each category, `hue` displays all groups together using different colors.

### 📌 Why Use `hue`?

- 🎨 Differentiate categories using colors.
- 📊 Compare multiple groups in a single visualization.
- 👀 Identify patterns, trends, and differences between categories.
- 📈 Improve the readability of complex datasets.
- 📄 Create informative and professional visualizations.

### 🌍 Real-World Example

A restaurant owner may use **`hue="sex"`** to compare spending patterns between male and female customers, or **`hue="smoker"`** to compare the total bill of smokers and non-smokers in the same chart.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Create a scatter plot with colors based on dining time
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)

# Add a descriptive title
plt.title("Scatter Plot Using Hue")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **`hue="time"`** parameter uses different colors to distinguish between **Lunch** and **Dinner** customers.
- The legend is automatically generated, making it easy to identify each dining time.
- **Dinner** transactions are more numerous than **Lunch** transactions in the dataset.
- Dinner customers are spread across a **wider range of total bill and tip values**, including the highest bills and tips.
- Lunch transactions are mostly concentrated between **$10–25** for the total bill and **$1–4** for the tip.
- Both Lunch and Dinner groups show a **positive relationship** between **Total Bill** and **Tip**, indicating that customers generally leave larger tips as their total bill increases.
- Using **`hue`** makes it easier to compare multiple categories within the same visualization without creating separate plots.

> **💡 Note:** The `hue` parameter automatically assigns a unique color to each category and creates a legend. This helps compare groups more effectively while keeping all observations in a single chart.

# 📏 Using `size`

The **`size`** parameter changes the size of markers based on the values of a **numerical variable**. Larger values are represented by larger markers, while smaller values appear as smaller markers.

This allows an additional dimension of information to be displayed within the same visualization without creating extra charts.

### 📌 Why Use `size`?

- 📊 Represent a numerical variable using marker size.
- 👀 Display more information in a single visualization.
- 📈 Identify patterns involving three or more variables.
- 🎨 Enhance Exploratory Data Analysis (EDA) by adding another visual dimension.
- 📄 Create more informative and professional scatter plots.

### 🌍 Real-World Example

A sales analyst might use:
- **x-axis** → Sales Revenue
- **y-axis** → Profit
- **size** → Number of Customers

This allows three variables to be analyzed simultaneously in a single chart.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Create a scatter plot with marker sizes based on party size
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    size="size",
    sizes=(40, 250)
)

# Add a descriptive title
plt.title("Scatter Plot Using Size")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **`size`** parameter changes the marker size based on the **party size** (`size`) variable.
- The legend shows that **larger circles represent larger party sizes**, while **smaller circles represent smaller parties**.
- The scatter plot shows a **positive relationship** between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- Larger markers are mostly concentrated in the **higher total bill** region, suggesting that **larger groups tend to have higher restaurant bills**.
- Smaller party sizes (1–2 people) are more common in the dataset, while larger parties (5–6 people) occur less frequently.
- Using marker size allows an additional numerical variable to be visualized without creating another chart.

> **💡 Note:** The `size` parameter is best used with **numerical variables**. The `sizes=(40, 250)` argument controls the minimum and maximum marker sizes, making differences between values easier to recognize.

# ✨ Using `style`

The **`style`** parameter changes the **marker shape** (or line style) based on a **categorical variable**. It allows different groups to be distinguished using symbols instead of colors.

This is especially useful when visualizations are printed in black and white or when colors alone are not enough to differentiate categories.

### 📌 Why Use `style`?

- 🔷 Differentiate categories using marker shapes.
- 👀 Improve readability when multiple groups are displayed.
- 🖨️ Create visualizations that remain understandable in grayscale or printed reports.
- 📊 Compare categorical groups without relying only on color.
- 🎨 Combine with `hue` and `size` to display multiple variables in a single chart.

### 🌍 Real-World Example

A restaurant analyst may use:

- **X-axis** → Total Bill
- **Y-axis** → Tip
- **Style** → Customer Gender

This allows male and female customers to be distinguished using different marker shapes, even if the visualization is viewed without color.

```python
# Create a figure with a custom size
plt.figure(figsize=(8, 5))

# Create a scatter plot with different marker styles based on smoker status
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    style="smoker"
)

# Add a descriptive title
plt.title("Scatter Plot Using Style")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- The **`style="smoker"`** parameter uses **different marker shapes** to distinguish between **Smokers (Yes)** and **Non-Smokers (No)**.
- The legend clearly identifies which marker shape corresponds to each customer group.
- Both smokers and non-smokers are distributed across a wide range of **total bill** and **tip** values.
- The scatter plot shows a **positive relationship** between **Total Bill** and **Tip** for both groups, indicating that customers generally leave higher tips as their total bill increases.
- The marker shapes make it easy to compare the two groups without relying solely on colors.
- Using `style` adds an additional visual distinction while preserving the readability of the chart.

> **💡 Note:** The `style` parameter is particularly useful for making visualizations accessible. Different marker shapes allow readers to distinguish categories even in grayscale prints or when colors are difficult to differentiate.

# 🎯 Combining `hue`, `style`, and `size`

Seaborn allows multiple **semantic mappings** to be used within a single visualization. By combining **color (`hue`)**, **marker shape (`style`)**, and **marker size (`size`)**, you can represent several variables at the same time.

This creates richer and more informative visualizations without requiring multiple separate charts.

### 📌 Why Combine Semantic Mappings?

- 📊 Visualize multiple variables in a single chart.
- 👀 Identify complex patterns and relationships more easily.
- 📈 Perform more effective Exploratory Data Analysis (EDA).
- 🎨 Improve storytelling by displaying additional information.
- 📄 Create professional-quality visualizations for reports and dashboards.

### 🌍 Real-World Example

A restaurant analyst may create a scatter plot where:

- **X-axis** → Total Bill
- **Y-axis** → Tip
- **Color (`hue`)** → Dining Time (Lunch/Dinner)
- **Marker Shape (`style`)** → Smoker Status
- **Marker Size (`size`)** → Party Size

This single visualization provides insights into **five variables simultaneously**, making it much more informative than a basic scatter plot.

```python
# Create a figure with a custom size
plt.figure(figsize=(10, 6))

# Create a scatter plot using multiple semantic mappings
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    style="smoker",
    size="size",
    sizes=(40,250)
)

# Add a descriptive title
plt.title("Combining Hue, Style, and Size")

# Label the axes
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

### 📌 Observation

- **Color** distinguishes customers by **Dining Time** (**Lunch** and **Dinner**).
- **Marker shape** differentiates **Smokers** and **Non-Smokers**.
- **Marker size** represents the **Party Size**, where larger circles indicate larger groups.
- The scatter plot shows a **positive relationship** between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- **Dinner** customers appear more frequently and are spread across a wider range of total bill and tip values than **Lunch** customers.
- Larger party sizes are generally associated with **higher total bills**, as indicated by the larger markers appearing toward the right side of the chart.
- Combining multiple semantic mappings provides a comprehensive view of the dataset while allowing several variables to be compared simultaneously.

> **💡 Note:** This visualization represents **five variables** in a single chart:
>
> - **X-axis:** Total Bill
> - **Y-axis:** Tip
> - **Color (`hue`):** Dining Time
> - **Marker Shape (`style`):** Smoker Status
> - **Marker Size (`size`):** Party Size
>
> While combining multiple semantic mappings is powerful, avoid adding unnecessary variables, as overly complex visualizations can become difficult to interpret.

# 🪟 FacetGrid

`FacetGrid` creates a grid of plots by splitting the dataset based on one or more **categorical variables**. Each subplot displays a subset of the data, making it easy to compare patterns across different groups.

Instead of combining all observations into a single chart, `FacetGrid` creates **multiple smaller charts** with the same axes and scales, allowing for clearer comparisons.

### 📌 Why Use `FacetGrid`?

- 📊 Compare multiple categories side by side.
- 👀 Identify differences and similarities between groups.
- 📈 Explore patterns that may be hidden in a single plot.
- 🎨 Create clean and organized multi-panel visualizations.
- 📄 Perform effective Exploratory Data Analysis (EDA).

### 🌍 Real-World Example

A restaurant analyst may create separate scatter plots for **Lunch** and **Dinner** customers to compare the relationship between **Total Bill** and **Tip**. Viewing each group separately makes it easier to identify differences in customer spending behavior.

```python
# Create a FacetGrid based on dining time
g = sns.FacetGrid(
    tips,
    col="time",
    height=5,
    aspect=1
)

# Map a scatter plot onto each facet
g.map(
    sns.scatterplot,
    "total_bill",
    "tip"
)

# Add titles and axis labels
g.set_titles("{col_name}")
g.set_axis_labels("Total Bill ($)", "Tip ($)")

# Add an overall title
g.fig.suptitle(
    "FacetGrid: Total Bill vs Tip by Dining Time",
    fontsize=16
)

# Adjust layout for the title
g.fig.tight_layout()
g.fig.subplots_adjust(top=0.85)

# Display the plots
plt.show()
```

### 📌 Observation

- `FacetGrid` separates the dataset into **two scatter plots** based on the **Dining Time** categories: **Lunch** and **Dinner**.
- Both subplots use the **same axis scales**, making it easy to compare the relationship between **Total Bill** and **Tip** across the two groups.
- Both **Lunch** and **Dinner** customers show a **positive relationship** between **Total Bill** and **Tip**, indicating that customers generally leave higher tips as their total bill increases.
- The **Dinner** subplot contains noticeably **more observations** than the **Lunch** subplot, suggesting that most transactions in the dataset occur during dinner.
- Dinner customers also exhibit a **wider range of total bill and tip values**, including the highest bills and tips in the dataset.
- Lunch customers are mainly concentrated between **$10–25** for the total bill and **$1–4** for the tip, showing a comparatively narrower distribution.
- By separating the data into multiple panels, `FacetGrid` makes it easier to compare customer behavior across dining times without overlapping points.

> **💡 Note:** `FacetGrid` creates multiple plots with **shared axis scales** by default. This ensures that differences between groups reflect the actual data rather than changes in the axis ranges, making comparisons more reliable.

# 📊 FacetGrid with Multiple Categories

`FacetGrid` can create a grid of plots using both **rows** and **columns** simultaneously. This allows the dataset to be split by **two categorical variables**, making it easier to compare multiple groups at the same time.

Each subplot represents a unique combination of the selected row and column categories while maintaining consistent axis scales for fair comparisons.

### 📌 Why Use Multiple Categories?

- 📊 Compare two categorical variables simultaneously.
- 👀 Identify patterns across different groups.
- 📈 Discover trends that may be hidden in a single visualization.
- 🎨 Create well-organized multi-panel visualizations.
- 📄 Perform deeper Exploratory Data Analysis (EDA).

### 🌍 Real-World Example

A restaurant analyst may compare customer spending by:

- **Columns (`col`)** → Dining Time (Lunch / Dinner)
- **Rows (`row`)** → Smoker Status (Yes / No)

This creates separate plots for each combination, making it easy to compare customer behavior across both dining time and smoking status.

```python
# Create a FacetGrid using rows and columns
g = sns.FacetGrid(
    tips,
    row="sex",
    col="time",
    height=4,
    aspect=1
)

# Map a scatter plot onto each facet
g.map(
    sns.scatterplot,
    "total_bill",
    "tip"
)

# Add titles and axis labels
g.set_titles("{row_name} | {col_name}")
g.set_axis_labels("Total Bill ($)", "Tip ($)")

# Add an overall title
g.fig.suptitle(
    "FacetGrid: Total Bill vs Tip by Gender and Dining Time",
    fontsize=16
)

# Adjust layout
g.fig.tight_layout()
g.fig.subplots_adjust(top=0.90)

# Display the plots
plt.show()
```

### 📌 Observation

- The `FacetGrid` divides the dataset into **four subplots** based on **Gender** (rows) and **Dining Time** (columns).
- Each subplot represents one unique combination:
  - **Male | Lunch**
  - **Male | Dinner**
  - **Female | Lunch**
  - **Female | Dinner**
- All four subplots show a **positive relationship** between **Total Bill** and **Tip**, indicating that customers generally leave higher tips as their total bill increases.
- The **Male | Dinner** subplot contains the **largest number of observations** and covers the widest range of total bill and tip values.
- The **Lunch** subplots (both Male and Female) contain fewer observations and are mostly concentrated at **lower total bill values**.
- The **Dinner** subplots display higher total bills and higher tips compared to the Lunch subplots, suggesting that dinner customers generally spend more.
- Male and female customers exhibit **similar overall tipping patterns**, with differences mainly in the number of transactions rather than the relationship between bill amount and tip.
- Splitting the data into multiple panels makes it much easier to compare customer behavior across **both Gender and Dining Time** without overlapping points.

> **💡 Note:** `FacetGrid` keeps the **same axis scales** across all subplots by default. This ensures that differences between groups reflect the actual data rather than changes in the chart scale, making comparisons fair and reliable.

# 🔥 PairGrid

`PairGrid` is a more flexible and customizable version of Seaborn's **`pairplot()`**. It creates a grid of pairwise relationships between numerical variables while allowing **different plot types** to be displayed in different parts of the grid.

Unlike `pairplot()`, where the same plot type is used throughout, `PairGrid` gives you full control over the **upper triangle**, **lower triangle**, and **diagonal**.

### 📌 Why Use `PairGrid`?

- 📊 Customize pairwise visualizations.
- 👀 Display different plot types in different sections of the grid.
- 📈 Explore relationships between multiple numerical variables.
- 🎨 Create more informative and professional EDA visualizations.
- 🔍 Identify correlations, distributions, clusters, and outliers more effectively.

### 🌍 Real-World Example

A restaurant analyst may use `PairGrid` to analyze the relationship between:

- **Total Bill**
- **Tip**
- **Party Size**

For example:

- **Upper Triangle** → Scatter Plots
- **Lower Triangle** → KDE Plots
- **Diagonal** → Histograms

This provides a richer understanding of the dataset than a standard `pairplot()`.

---

## 💡 Note

> **💡 Note:** `PairGrid` is ideal when you need different visualization types for different parts of the grid. If you simply want the default pairwise relationships, `pairplot()` is usually faster and requires less code.

---

## 🎯 PairPlot vs PairGrid

| Feature | `pairplot()` | `PairGrid` |
|----------|--------------|------------|
| Ease of Use | ⭐⭐⭐⭐⭐ | ⭐⭐⭐ |
| Customization | Limited | Extensive |
| Different Plot Types | ❌ No | ✅ Yes |
| Best For | Quick EDA | Advanced EDA |
| Code Complexity | Low | Moderate |

---

## ⭐ Professional Tip

A `PairGrid` is divided into three sections:

| Section | Purpose | Common Plot |
|----------|----------|-------------|
| **Upper Triangle** | Compare variable pairs | Scatter Plot |
| **Lower Triangle** | Compare variable pairs | KDE Plot or Regression Plot |
| **Diagonal** | Show individual distributions | Histogram or KDE Plot |

This flexibility makes `PairGrid` one of the most powerful tools in Seaborn for advanced Exploratory Data Analysis (EDA).

```python
# Create a PairGrid using selected numerical variables
g = sns.PairGrid(
    tips,
    vars=["total_bill", "tip", "size"]
)

# Scatter plots in the upper triangle
g.map_upper(sns.scatterplot)

# KDE plots in the lower triangle
g.map_lower(
    sns.kdeplot,
    fill=True
)

# Histograms on the diagonal
g.map_diag(sns.histplot)

# Add an overall title
g.fig.suptitle(
    "PairGrid: Relationship Between Total Bill, Tip, and Party Size",
    fontsize=16
)

# Adjust layout
g.fig.tight_layout()
g.fig.subplots_adjust(top=0.95)

# Display the plots
plt.show()
```

### 📌 Observation

- The `PairGrid` visualizes the pairwise relationships between the numerical variables **Total Bill**, **Tip**, and **Party Size**.
- The **diagonal** contains **histograms**, showing the distribution of each variable individually.
- The **upper triangle** uses **scatter plots**, making it easy to observe relationships between pairs of variables.
- The **lower triangle** uses **KDE (Kernel Density Estimation) plots**, highlighting where observations are most densely concentrated.
- There is a **strong positive relationship** between **Total Bill** and **Tip**, indicating that customers with higher bills generally leave higher tips.
- **Party Size** also shows a positive relationship with **Total Bill**, suggesting that larger groups tend to spend more.
- The relationship between **Party Size** and **Tip** is moderately positive, meaning larger groups often leave higher tips, although the relationship is not as strong as that between Total Bill and Tip.
- The histograms reveal that:
  - **Total Bill** is **right-skewed**, with most bills being relatively small and only a few very large bills.
  - **Tip** is also **right-skewed**, with most tips between approximately **$2 and $4**.
  - **Party Size** is concentrated around **2 people**, making it the most common group size in the dataset.
- The KDE plots show that the highest density of observations occurs for:
  - **Total Bill:** approximately **$10–25**
  - **Tip:** approximately **$2–4**
  - **Party Size:** **2–3 people**
- Combining different visualization types in one grid provides a more comprehensive understanding of the dataset than using a single plot type.

> **💡 Note:** `PairGrid` allows different plot types to be assigned to different sections of the grid. This flexibility helps analyze both **relationships** (scatter plots), **distributions** (histograms), and **data density** (KDE plots) within a single visualization.

# 📊 JointGrid

`JointGrid` provides complete control over **joint** and **marginal** distributions.

Unlike `jointplot()`, which automatically creates both plots using a predefined style, `JointGrid` allows you to customize the **main plot** and the **marginal plots** independently.

This makes it one of the most flexible visualization tools in Seaborn for analyzing the relationship between two numerical variables.

### 📌 Why Use `JointGrid`?

- 📊 Customize the central relationship plot.
- 📈 Customize the marginal distributions separately.
- 👀 Analyze relationships and distributions simultaneously.
- 🎨 Combine different plot types in a single visualization.
- 📄 Perform detailed Exploratory Data Analysis (EDA).

### 🌍 Real-World Example

A restaurant analyst may want to study the relationship between **Total Bill** and **Tip** by displaying:

- **Joint Plot (Center)** → Scatter Plot
- **Top Marginal Plot** → Histogram of Total Bill
- **Right Marginal Plot** → Histogram of Tip

This provides both the relationship between the variables and the distribution of each variable in one visualization.

```python
# Create a JointGrid
g = sns.JointGrid(
    data=tips,
    x="total_bill",
    y="tip",
    height=6
)

# Plot a scatter plot in the center
# and histograms on the margins
g.plot(
    sns.scatterplot,
    sns.histplot
)

# Add axis labels
g.set_axis_labels(
    "Total Bill ($)",
    "Tip ($)"
)

# Add an overall title
g.figure.suptitle(
    "JointGrid: Total Bill vs Tip",
    fontsize=16
)

# Adjust layout
g.figure.tight_layout()
g.figure.subplots_adjust(top=0.95)

# Display the plot
plt.show()
```

### 📌 Observation

- The **center scatter plot** shows a **strong positive relationship** between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- Most observations are concentrated between **$10–25** for the **Total Bill** and **$2–4** for the **Tip**, suggesting that these are the most common transaction values.
- A few points appear in the upper-right corner of the scatter plot, representing customers with **very high bills and correspondingly high tips**.
- The **top histogram** shows the distribution of **Total Bill**, which is **right-skewed**. Most bills are relatively small, while only a few customers have very high bills.
- The **right histogram** shows the distribution of **Tip**, which is also **right-skewed**, with most tips ranging between **$2 and $4**.
- The scatter plot does not indicate any strong negative relationship or distinct clusters; instead, it shows a consistent upward trend across the dataset.
- By combining the scatter plot with marginal histograms, `JointGrid` allows both the **relationship between variables** and their **individual distributions** to be analyzed simultaneously.

> **💡 Note:** `JointGrid` combines two types of analysis in a single visualization:
>
> - **Joint Plot (Center):** Shows the relationship between two numerical variables.
> - **Marginal Plots (Top & Right):** Show the distribution of each variable individually.
>
> This makes it easier to understand both **correlation** and **distribution** without creating multiple separate charts.

# 🎨 Combining Seaborn with Matplotlib

Seaborn is built on top of **Matplotlib**, which means both libraries work seamlessly together.

While Seaborn simplifies the creation of beautiful statistical visualizations, Matplotlib provides detailed control over the appearance of the plots.

By combining both libraries, you can create visualizations that are both attractive and highly customizable.

### 📌 Why Combine Seaborn with Matplotlib?

- 📝 Add informative titles and axis labels.
- 📏 Control figure size and layout.
- 🎨 Customize colors, fonts, and styles.
- 📍 Add annotations to highlight important insights.
- 🏷️ Customize legends and their positions.
- 📐 Modify grids, ticks, and axis limits.
- 📄 Create publication-quality charts for reports and presentations.

### 🌍 Real-World Example

A data analyst may use **Seaborn** to create a scatter plot and then use **Matplotlib** to:

- Add a descriptive title.
- Label the axes.
- Adjust the figure size.
- Customize the legend.
- Highlight important observations with annotations.

This combination results in professional and easy-to-understand visualizations.

```python
# Create a figure with a custom size
plt.figure(figsize=(10, 6))

# Create a Seaborn scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    palette="Set2"
)

# Add a customized title
plt.title(
    "Total Bill vs Tip",
    fontsize=16,
    fontweight="bold"
)

# Add axis labels
plt.xlabel("Total Bill ($)", fontsize=12)
plt.ylabel("Tip ($)", fontsize=12)

# Add a customized grid
plt.grid(alpha=0.3)

# Adjust layout automatically
plt.tight_layout()

# Display the plot
plt.show()
```

### 📌 Observation

- The scatter plot shows the relationship between **Total Bill** and **Tip**, with each point colored according to the **Day** of the week.
- A **positive relationship** exists between **Total Bill** and **Tip**, indicating that customers who spend more generally leave higher tips.
- The **Set2** color palette clearly distinguishes transactions across **Thursday**, **Friday**, **Saturday**, and **Sunday**, making group comparisons easier.
- **Saturday** and **Sunday** contain more observations than **Thursday** and **Friday**, suggesting that the restaurant is busier on weekends.
- Most transactions are concentrated between **$10–25** for the **Total Bill** and **$2–4** for the **Tip**, representing the most common customer spending pattern.
- A few observations with **very high bills** (around **$45–50**) also have **higher tips**, indicating occasional high-value transactions.
- The customized title, axis labels, and subtle grid lines improve the readability and overall presentation of the visualization.
- The legend makes it easy to identify which color corresponds to each day, allowing quick comparisons across different days of the week.

> **💡 Note:** This visualization demonstrates how **Seaborn** and **Matplotlib** work together. Seaborn creates the statistical visualization, while Matplotlib enhances it with customized titles, labels, grids, figure sizing, and layout adjustments.

- 👥 Design visualizations that are easy for your audience to understand, even if they are unfamiliar with the dataset.

# ❌ Common Mistakes

Avoid these common mistakes when customizing Seaborn visualizations:

- 🎨 Overusing colors or using too many color palettes in a single chart.
- 📊 Displaying too many variables in one visualization, making it difficult to interpret.
- 🌈 Choosing inappropriate color palettes with poor contrast or colors that are difficult to distinguish.
- 🏷️ Forgetting to add clear titles, axis labels, or legends.
- 🔍 Ignoring readability by using small fonts, overlapping labels, or cluttered layouts.
- 📍 Placing legends where they hide important data points.
- 📏 Creating figures that are too small or too large for the intended audience.
- 🎯 Using semantic mappings (`hue`, `style`, or `size`) when they do not provide meaningful insights.
- 📈 Selecting the wrong chart type for the data being analyzed.
- 🚫 Drawing conclusions from a single visualization without validating them using other plots.
- ❓ Creating visualizations without a clear analytical purpose.
