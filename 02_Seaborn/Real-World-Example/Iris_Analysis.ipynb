# 🌸 Species Distribution

Before analyzing numerical features, let's understand how the samples are distributed across different iris species.

This visualization helps us analyze:

- Number of samples in each species.
- Dataset balance.
- Class distribution.
- Whether the dataset is suitable for machine learning.

```python
# This count plot shows the number of flowers
# belonging to each iris species.

plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x="species"
)

# Add value labels
for container in ax.containers:
    ax.bar_label(
        container,
        fontsize=10,
        fontweight="bold"
    )

plt.title(
    "Distribution of Iris Species",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel(
    "Species",
    fontsize=12
)

plt.ylabel(
    "Number of Flowers",
    fontsize=12
)

plt.grid(axis="y", alpha=0.3)

plt.tight_layout()

plt.show()
```

### 📌 Observation

- The dataset contains **149 iris flowers** distributed across three species.
- **Setosa** and **Versicolor** each contain **50 samples**, while **Virginica** contains **49 samples** after removing duplicate records.
- The dataset remains **well-balanced**, with nearly equal representation of all three species.
- The small difference of one sample is unlikely to affect the overall analysis or model performance.
- Overall, the balanced distribution makes this dataset suitable for classification and comparative analysis.

### 💡 Insight

The nearly balanced class distribution minimizes the risk of model bias toward any single species. This makes the Iris dataset an excellent choice for learning classification algorithms and evaluating visualization techniques.

# 📏 Sepal Length Distribution

Sepal Length is one of the primary measurements used to distinguish iris species.

This visualization helps us analyze:

- Distribution of Sepal Length.
- Most common values.
- Spread of the data.
- Presence of skewness or outliers.

```python
# This histogram shows the distribution
# of Sepal Length.

plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="sepal_length",
    bins=20,
    kde=True,
    edgecolor="black"
)

plt.axvline(
    df["sepal_length"].mean(),
    color="red",
    linestyle="--",
    linewidth=2,
    label="Mean"
)

plt.axvline(
    df["sepal_length"].median(),
    color="green",
    linestyle="--",
    linewidth=2,
    label="Median"
)

plt.title(
    "Distribution of Sepal Length",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel(
    "Sepal Length (cm)",
    fontsize=12
)

plt.ylabel(
    "Frequency",
    fontsize=12
)

plt.grid(alpha=0.3)

plt.legend()

plt.tight_layout()

plt.show()
```

### 📌 Observation

- Most flowers have a **sepal length between 5 cm and 7 cm**, indicating that medium-sized sepals are the most common.
- The distribution appears **approximately normal** with a slight right skew.
- The **mean and median** are very close, suggesting minimal skewness and a fairly symmetric distribution.
- Very short and very long sepal lengths occur less frequently.
- Overall, Sepal Length exhibits moderate variability across the iris dataset.

### 💡 Insight

Sepal Length provides useful information about flower morphology, but it is not sufficient on its own to clearly distinguish all three iris species. Additional features, particularly petal measurements, are expected to offer stronger discrimination.

# 📏 Sepal Width Distribution

Sepal Width represents another important morphological feature of iris flowers. Understanding its distribution helps identify the variability of flower dimensions across the dataset.

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="sepal_width",
    bins=20,
    kde=True,
    edgecolor="black"
)

plt.axvline(
    df["sepal_width"].mean(),
    color="red",
    linestyle="--",
    linewidth=2,
    label="Mean"
)

plt.axvline(
    df["sepal_width"].median(),
    color="green",
    linestyle="--",
    linewidth=2,
    label="Median"
)

plt.title(
    "Distribution of Sepal Width",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel("Sepal Width (cm)")

plt.ylabel("Frequency")

plt.grid(alpha=0.3)

plt.legend()

plt.tight_layout()

plt.show()
```

### 📌 Observation

- Most flowers have a **sepal width between 2.5 cm and 3.5 cm**, representing the most common range in the dataset.
- The distribution is approximately **bell-shaped** with only slight skewness.
- The **mean and median** are closely aligned, indicating a balanced distribution with minimal asymmetry.
- Extremely narrow and extremely wide sepals are relatively rare.
- Compared with Sepal Length, Sepal Width shows **less overall variation** among flowers.

### 💡 Insight

Sepal Width contributes valuable information for describing flower characteristics; however, similar to Sepal Length, it does not provide complete separation between species. Combining sepal measurements with petal features results in much better species differentiation.

# 📦 Sepal Length by Species

Box plots provide a statistical summary of numerical features across different species. They display the median, interquartile range (IQR), spread, and potential outliers.

This visualization helps us compare Sepal Length across the three iris species.

```python
# Compare Sepal Length across species

plt.figure(figsize=(9,6))

sns.boxplot(
    data=df,
    x="species",
    y="sepal_length"
)

plt.title(
    "Sepal Length by Species",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel(
    "Species",
    fontsize=12
)

plt.ylabel(
    "Sepal Length (cm)",
    fontsize=12
)

plt.grid(axis="y", alpha=0.3)

plt.tight_layout()

plt.show()
```

### 📌 Observation

- **Virginica** exhibits the highest median Sepal Length, indicating generally larger flowers.
- **Versicolor** falls between Setosa and Virginica in terms of sepal length.
- Virginica shows greater variability in Sepal Length and contains a few potential outliers.
- Although the medians differ, there is some overlap between Versicolor and Virginica.

### 💡 Insight

Sepal Length helps distinguish the species to some extent, especially Setosa from the others. However, the overlap between Versicolor and Virginica suggests that additional features are needed for accurate classification.

# 🌈 Pair Plot

A pair plot visualizes pairwise relationships among numerical features while simultaneously displaying the distribution of each feature. Coloring points by species helps identify natural clusters and feature separability.

This visualization helps us analyze:

- Relationships between numerical features.
- Distribution of each feature.
- Species clustering.
- Feature combinations useful for classification.

```python
sns.pairplot(
    df,
    hue="species",
    diag_kind="hist",
    height=2.8
)

plt.suptitle(
    "Pairwise Relationships Among Iris Features",
    fontsize=16,
    fontweight="bold",
    y=1.02
)

plt.show()
```

### 📌 Observation

- **Setosa** forms a completely distinct cluster across most feature combinations, making it easily distinguishable.
- **Versicolor** and **Virginica** show slight overlap in sepal-related features but become much more separable when petal measurements are considered.
- **Petal Length** and **Petal Width** exhibit a **strong positive relationship**, indicating that flowers with longer petals generally have wider petals.
- **Sepal Length** also shows a moderate positive relationship with petal measurements.
- **Sepal Width** demonstrates weaker relationships with the other numerical features.
- Histograms along the diagonal reveal that petal-related features produce more distinct distributions than sepal-related features.
- Overall, petal measurements provide significantly better species separation than sepal measurements.

### 💡 Insight

The pair plot clearly demonstrates that **Petal Length** and **Petal Width** are the most powerful features for distinguishing iris species. Setosa can be separated almost perfectly using these features, while Versicolor and Virginica show only slight overlap. These relationships explain why the Iris dataset is widely used as a benchmark for classification algorithms.

# 🔥 Correlation Heatmap

Correlation analysis helps us understand the strength and direction of relationships between numerical features.

A correlation coefficient ranges from **-1 to +1**, where:

- **+1** → Strong Positive Correlation
- **0** → No Correlation
- **-1** → Strong Negative Correlation

This visualization helps identify which features are most closely related and provides useful insights for feature selection in machine learning.

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include="number")

# Compute correlation matrix
correlation = numerical_df.corr()

plt.figure(figsize=(9,7))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    square=True
)

plt.title(
    "Correlation Heatmap of Iris Features",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.tight_layout()

plt.show()
```

### 📌 Observation

- **Petal Length** and **Petal Width** have the strongest positive correlation (**0.96**), indicating that flowers with longer petals generally have wider petals.
- **Sepal Length** also shows a strong positive correlation with both **Petal Length (0.87)** and **Petal Width (0.82)**.
- **Sepal Width** has a weak negative correlation with the other numerical features, particularly with **Petal Length (-0.43)** and **Petal Width (-0.36)**.
- The strongest relationships are observed among **petal-related features**, while sepal-related features exhibit weaker associations.
- Overall, petal measurements contain the most informative relationships within the Iris dataset.

### 💡 Insight

The correlation heatmap confirms that **Petal Length** and **Petal Width** are the most important features for distinguishing iris species. Their strong positive relationship suggests they carry similar information and play a crucial role in classification models. In contrast, **Sepal Width** contributes additional but less correlated information, making it a complementary feature rather than a primary predictor.

# 💡 Best Practices

When performing Exploratory Data Analysis (EDA), consider the following best practices:

- Understand the dataset before creating visualizations.
- Verify data quality by checking for missing values and duplicate records.
- Select visualization types that best represent the underlying data.
- Use clear titles, labels, legends, and annotations.
- Maintain consistent formatting throughout the notebook.
- Validate observations using multiple visualizations.
- Remember that **correlation does not imply causation**.
- Summarize findings with meaningful observations and actionable insights.

# ❌ Common Mistakes

Avoid the following mistakes:

- Ignoring missing values before analysis.
- Drawing conclusions from a single visualization.
- Using too many colors or plot elements.
- Forgetting titles, axis labels, or legends.
- Assuming correlation implies causation.
