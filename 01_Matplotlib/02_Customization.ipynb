````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

# 📐 Figure vs Axes

Matplotlib is built around two main objects:

- **Figure**: The entire window or canvas.
- **Axes**: The individual plotting area inside the figure.

A figure can contain one or more axes.

````python
fig, ax = plt.subplots(figsize=(8,5))

x = np.arange(1,11)
y = x**2

ax.plot(x,y)

ax.set_title("Figure and Axes Example")

plt.show()
````

## 📊 Creating Multiple Subplots

Subplots allow multiple charts to be displayed within the same figure.

This is useful for comparing different datasets or visualizations side by side.

````python
x = np.arange(1,11)

fig, ax = plt.subplots(1,2, figsize=(12,5))

ax[0].plot(x, x**2)
ax[0].set_title("Square")

ax[1].plot(x, x**3)
ax[1].set_title("Cube")

plt.tight_layout()

plt.show()
````

## 📏 Figure Size

The `figsize` parameter controls the width and height of the figure in inches.

A larger figure improves readability for presentations and reports.

````python
plt.figure(figsize=(10,6))

plt.plot(x, x**2)

plt.title("Large Figure")

plt.show()
````

## 🎨 Plot Styles

Matplotlib provides built-in styles that change the appearance of plots.

You can view all available styles using:

```python
plt.style.available
```

````python
print(plt.style.available)
````

````python
plt.style.use("ggplot")

plt.figure(figsize=(8,5))

plt.plot(x, x**2)

plt.title("GGPlot Style")

plt.show()
````

# 🌈 Colors in Matplotlib

Colors play an important role in making visualizations attractive and easy to understand.

Matplotlib supports multiple ways to specify colors:

- Named Colors (`"red"`, `"blue"`, `"green"`)
- Short Color Codes (`"r"`, `"g"`, `"b"`)
- Hexadecimal Colors (`"#FF5733"`)
- RGB Values

````python
x = np.arange(1,11)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, x**2, color="royalblue", linewidth=2)

ax.set_title("Custom Color Example")

plt.show()
````

# 🎨 Colormaps

A colormap is a collection of colors used to represent data.

Matplotlib provides many built-in colormaps such as:

- viridis
- plasma
- inferno
- magma
- cividis

These are especially useful for heatmaps, scatter plots, and image visualizations.

````python
print(plt.colormaps())
````

# 📏 Customizing Line Styles

Line styles help distinguish between different datasets.

Common styles include:

- Solid (`"-"`)
- Dashed (`"--"`)
- Dash-dot (`"-."`)
- Dotted (`":"`)

````python
x = np.arange(1,11)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(
    x,
    x**2,
    linestyle="--",
    linewidth=3,
    color="tomato"
)

ax.set_title("Dashed Line")

plt.show()
````

# 🔵 Marker Customization

Markers highlight individual observations.

Useful parameters include:

- marker
- markersize
- markerfacecolor
- markeredgecolor
- markeredgewidth

````python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(
    x,
    x**2,
    marker="o",
    markersize=10,
    markerfacecolor="gold",
    markeredgecolor="black",
    markeredgewidth=2
)

ax.set_title("Customized Markers")

plt.show()
````

# 📍 Setting Axis Limits

Sometimes we only want to display a specific range of values.

Use:

- `set_xlim()`
- `set_ylim()`

````python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, x**2)

ax.set_xlim(2,8)

ax.set_ylim(0,80)

ax.set_title("Custom Axis Limits")

plt.show()
````

# 🔢 Customizing Tick Locations

Ticks represent the values displayed on the axes.

You can manually specify tick positions for better control.

````python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, x**2)

ax.set_xticks(np.arange(1,11,2))

ax.set_yticks(np.arange(0,101,20))

ax.set_title("Custom Tick Positions")

plt.show()
````

# 🔄 Rotating Tick Labels

Rotating tick labels improves readability, especially when labels are long or closely spaced.

````python
months = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June"
]

sales = [15,18,25,30,28,35]

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(months, sales, marker="o")

plt.xticks(rotation=45)

plt.show()
````

# 🟦 Customizing Grid

Grid lines improve readability by making it easier to compare values across the chart.

Matplotlib allows you to customize:

- Grid color
- Line style
- Line width
- Transparency (alpha)

````python
x = np.arange(1,11)
y = x**2

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y)

ax.grid(
    True,
    color="gray",
    linestyle="--",
    linewidth=0.8,
    alpha=0.7
)

ax.set_title("Customized Grid")

plt.show()
````

# 📖 Legend Customization

Legends help identify multiple plotted datasets.

You can customize:

- Location
- Font size
- Background color
- Border color
- Shadow

````python
x = np.arange(1,11)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, x**2, label="Square")
ax.plot(x, x**3, label="Cube")

ax.legend(
    loc="upper left",
    fontsize=11,
    frameon=True,
    shadow=True
)

plt.show()
````

# 📝 Text Annotation

Annotations allow you to highlight important data points directly on a chart.

Use:

```python
ax.annotate()
```

This is commonly used in reports and presentations.

````python
x = np.arange(1,11)
y = x**2

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y, marker="o")

ax.annotate(
    "Maximum Value",
    xy=(10,100),
    xytext=(7,80),
    arrowprops=dict(
        arrowstyle="->",
        color="red"
    )
)

ax.set_title("Annotation Example")

plt.show()
````

# ➡️ Adding Arrow Annotations

Arrows are useful for drawing attention to significant events or observations in your visualization.

````python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y)

ax.annotate(
    "",
    xy=(8,64),
    xytext=(5,90),
    arrowprops=dict(
        arrowstyle="->",
        lw=2,
        color="blue"
    )
)

plt.show()
````

# 🔲 Adding Shapes

Matplotlib can draw geometric shapes such as:

- Rectangle
- Circle
- Polygon

These are useful for highlighting specific regions in a chart.

````python
from matplotlib.patches import Rectangle
````

````python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y)

rect = Rectangle(
    (3,10),
    width=3,
    height=30,
    fill=False,
    edgecolor="red",
    linewidth=2
)

ax.add_patch(rect)

ax.set_title("Rectangle Highlight")

plt.show()
````

# 🔵 Adding Circle

Circles can be used to emphasize a particular region or point in a visualization.

````python
from matplotlib.patches import Circle
````

````python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, y)

circle = Circle(
    (8,64),
    radius=1,
    fill=False,
    edgecolor="green",
    linewidth=2
)

ax.add_patch(circle)

ax.set_title("Circle Highlight")

plt.show()
````

# 💾 Saving Figures

You can save a visualization in various formats, including:

- PNG
- JPG

Use:

plt.savefig()

Saving figures is useful when preparing reports, presentations, or publications.

# 🌟 Professional Visualization Example

Let's combine the customization techniques learned in this notebook to create a clean and professional visualization.

This example includes:

- Figure customization
- Colors
- Line styles
- Markers
- Grid
- Legend
- Axis labels
- Title
- Annotation

````python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug"]
sales = [25,30,35,40,38,45,50,55]

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(
    months,
    sales,
    color="royalblue",
    linewidth=3,
    linestyle="-",
    marker="o",
    markersize=8,
    label="Monthly Sales"
)

ax.set_title(
    "Monthly Sales Trend",
    fontsize=18,
    fontweight="bold"
)

ax.set_xlabel("Month", fontsize=12)

ax.set_ylabel("Sales (in Thousands)", fontsize=12)

ax.grid(
    True,
    linestyle="--",
    alpha=0.6
)

ax.legend()

ax.annotate(
    "Highest Sales",
    xy=("Aug",55),
    xytext=("Jun",60),
    arrowprops=dict(
        arrowstyle="->",
        color="red",
        lw=2
    )
)

plt.tight_layout()

plt.show()
````

# 💡 Best Practices

When creating visualizations:

- Use meaningful titles.
- Label both axes clearly.
- Choose colors that are easy to distinguish.
- Avoid unnecessary decorations.
- Keep the design simple and clean.
- Use legends only when needed.
- Highlight important insights using annotations.
- Ensure the figure size is appropriate for the audience.
- Maintain consistent styling across all visualizations.

# ❌ Common Mistakes

Avoid the following mistakes:

- Missing chart titles.
- Unlabeled axes.
- Using too many colors.
- Overcrowding the visualization.
- Choosing inappropriate chart types.
- Ignoring readability.
- Using legends when only one dataset is plotted.
- Adding unnecessary grid lines.
