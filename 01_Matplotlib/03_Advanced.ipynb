# 📐 Advanced Figure Layouts

Complex dashboards often require multiple plots arranged in different layouts.

Matplotlib provides:

- `plt.subplot()`
- `plt.subplots()`
- `GridSpec`

For flexible layouts, `GridSpec` is the preferred approach.

````python
from matplotlib.gridspec import GridSpec
````

````python
fig = plt.figure(figsize=(10,6))

gs = GridSpec(2,2)

ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])

x = np.arange(1,11)

ax1.plot(x, x)
ax1.set_title("Top Plot")

ax2.plot(x, x**2)
ax2.set_title("Square")

ax3.plot(x, x**3)
ax3.set_title("Cube")

plt.tight_layout()

plt.show()
````

# ⚖️ Twin Axes

Sometimes two variables have different scales but share the same x-axis.

Matplotlib allows us to create a second y-axis using:

```python
ax.twinx()
```

````python
x = np.arange(1,11)

sales = x*100
profit = x*20

fig, ax1 = plt.subplots(figsize=(8,5))

ax1.plot(
    x,
    sales,
    color="blue",
    label="Sales"
)

ax1.set_ylabel("Sales")

ax2 = ax1.twinx()

ax2.plot(
    x,
    profit,
    color="red",
    label="Profit"
)

ax2.set_ylabel("Profit")

plt.title("Sales vs Profit")

plt.show()
````

# 📊 Logarithmic Scale

A logarithmic scale is useful when data spans several orders of magnitude.

Instead of increasing linearly, the axis increases exponentially.

Use:

- `set_xscale("log")`
- `set_yscale("log")`

````python
x = np.arange(1,11)

y = np.exp(x)

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x,y)

ax.set_yscale("log")

ax.set_title("Logarithmic Scale")

plt.show()
````

# 📉 Error Bars

Error bars represent the uncertainty or variability in measurements.

They are commonly used in scientific experiments and statistical analysis.

````python
x = np.arange(1,6)

y = [5,7,6,8,9]

error = [0.5,0.4,0.6,0.3,0.5]

plt.figure(figsize=(8,5))

plt.errorbar(
    x,
    y,
    yerr=error,
    fmt='o-',
    capsize=5
)

plt.title("Error Bar Example")

plt.show()
````

# 🌱 Stem Plot

A Stem Plot displays data using vertical lines (stems) extending from a baseline to each data point.

It is commonly used for:

- Discrete signals
- Digital signal processing (DSP)
- Frequency analysis
- Scientific and engineering applications

````python
x = np.arange(1, 11)
y = [3, 7, 2, 8, 5, 9, 4, 6, 1, 8]

plt.figure(figsize=(8,5))

plt.stem(x, y)

plt.title("Stem Plot")

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
````

# 👣 Step Plot

A Step Plot displays changes occurring at specific intervals.

It is useful for:

- Inventory levels
- Digital signals
- Electricity billing
- Piecewise constant data

````python
x = np.arange(1,11)
y = [2,2,4,4,6,6,8,8,10,10]

plt.figure(figsize=(8,5))

plt.step(
    x,
    y,
    where="mid"
)

plt.title("Step Plot")

plt.grid(True)

plt.show()
````

# 🎨 Fill Between

The `fill_between()` function fills the area between a curve and the x-axis (or between two curves).

It is commonly used for:

- Confidence intervals
- Profit and loss regions
- Area highlighting

````python
x = np.arange(1,11)

y = x**2

plt.figure(figsize=(8,5))

plt.plot(x, y)

plt.fill_between(
    x,
    y,
    alpha=0.3
)

plt.title("Fill Between Example")

plt.show()
````

# 📚 Stack Plot

A Stack Plot visualizes how different categories contribute to a total over time.

Common applications include:

- Revenue analysis
- Budget allocation
- Population studies
- Resource utilization

````python
days = [1,2,3,4,5]

study = [2,3,4,5,4]

exercise = [1,1,1,1,1]

sleep = [7,7,6,7,8]

plt.figure(figsize=(8,5))

plt.stackplot(
    days,
    study,
    exercise,
    sleep,
    labels=["Study","Exercise","Sleep"]
)

plt.legend(loc="upper left")

plt.title("Daily Activities")

plt.show()
````

# 📍 Event Plot

An Event Plot displays the occurrence of events along a line.

Applications include:

- Earthquake occurrences
- Machine failures
- Medical events
- Sports analytics

````python
events = [
    np.random.rand(15),
    np.random.rand(20),
    np.random.rand(12)
]

plt.figure(figsize=(8,5))

plt.eventplot(events)

plt.title("Event Plot")

plt.show()
````

# 🔲 Spy Plot

A Spy Plot visualizes the sparsity pattern of a matrix.

It is useful for:

- Sparse matrices
- Machine Learning
- Network analysis
- Scientific computing

````python
matrix = np.random.randint(
    0,
    2,
    (10,10)
)

plt.figure(figsize=(6,6))

plt.spy(matrix)

plt.title("Spy Plot")

plt.show()
````

# 📅 Plotting Dates

Many real-world datasets contain date and time information.

Matplotlib works seamlessly with Pandas datetime objects, making it easy to visualize time-series data.

````python
dates = pd.date_range(
    start="2026-01-01",
    periods=10
)

sales = np.random.randint(
    20,
    60,
    10
)

plt.figure(figsize=(10,5))

plt.plot(
    dates,
    sales,
    marker="o"
)

plt.title("Sales Over Time")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
````

# 🌟 Professional Example

Let's combine multiple advanced Matplotlib features into a single visualization.

This example demonstrates:

- Figure and Axes
- Grid customization
- Multiple datasets
- Annotations
- Legends
- Professional styling

````python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug"]

sales = [25,28,32,40,38,45,52,60]
profit = [5,7,9,12,11,14,16,20]

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(
    months,
    sales,
    marker="o",
    linewidth=2.5,
    label="Sales"
)

ax.plot(
    months,
    profit,
    marker="s",
    linewidth=2.5,
    label="Profit"
)

ax.set_title(
    "Monthly Sales and Profit Analysis",
    fontsize=16,
    fontweight="bold"
)

ax.set_xlabel("Month")
ax.set_ylabel("Amount")

ax.grid(
    True,
    linestyle="--",
    alpha=0.5
)

ax.legend()

ax.annotate(
    "Highest Sales",
    xy=("Aug",60),
    xytext=("Jun",70),
    arrowprops=dict(
        arrowstyle="->",
        lw=2
    )
)

plt.tight_layout()

plt.show()
````

# 💡 Best Practices

Follow these practices to create effective visualizations:

- Keep charts simple and uncluttered.
- Use appropriate chart types for the data.
- Add informative titles and axis labels.
- Choose colors that improve readability.
- Use legends only when necessary.
- Highlight important insights with annotations.
- Avoid excessive text and decorations.
- Use consistent formatting across all charts.
- Ensure figures are readable on different screen sizes.

# ❌ Common Mistakes

Avoid these common visualization mistakes:

- Overloading a chart with too much information.
- Using misleading axis scales.
- Choosing poor color combinations.
- Forgetting axis labels.
- Using inappropriate chart types.
- Ignoring figure size.
- Overusing annotations.
- Displaying too many grid lines.
