````python
# Import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Display plots inside Jupyter Notebook
%matplotlib inline
````

## 🔍 Understanding pyplot

`matplotlib.pyplot` is a collection of functions that allows you to create and customize plots.

Common functions include:

- `plt.plot()` → Create line plots
- `plt.title()` → Add a title
- `plt.xlabel()` → Label the x-axis
- `plt.ylabel()` → Label the y-axis
- `plt.legend()` → Display legends
- `plt.grid()` → Show grid lines
- `plt.show()` → Display the plot

## 🚀 Your First Plot

Let's create our first line chart using Python lists.

````python
# Sample data

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 18, 22]

plt.plot(x, y)

plt.show()
````

### Explanation

The `plt.plot()` function connects the data points using straight lines.

Here:

- `x` represents the values on the horizontal axis.
- `y` represents the values on the vertical axis.

The `plt.show()` function displays the final plot.

## 📈 Plotting with NumPy Arrays

Matplotlib works seamlessly with NumPy arrays, making it ideal for scientific computing and data analysis.

````python
x = np.arange(1, 11)
y = x ** 2

plt.plot(x, y)

plt.show()
````

## 📐 Figure Size

The `figure()` function allows you to control the size of your visualization.

**Syntax:**

```python
plt.figure(figsize=(width, height))
```

The size is measured in **inches**.

Using an appropriate figure size improves readability, especially for presentations and dashboards.

````python
x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(8,5))

plt.plot(x, y)

plt.show()
````

## 🏷️ Adding Title and Axis Labels

A plot should always have a meaningful title and properly labeled axes.

Matplotlib provides:

- `plt.title()` → Adds a title
- `plt.xlabel()` → Labels the x-axis
- `plt.ylabel()` → Labels the y-axis

````python
x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(8,5))

plt.plot(x, y)

plt.title("Square Numbers")
plt.xlabel("Number")
plt.ylabel("Square")

plt.show()
````

## 🎨 Changing Line Color

The `color` parameter changes the color of the plotted line.

Examples:

- `"red"`
- `"blue"`
- `"green"`
- `"black"`
- `"orange"`
- Hex colors (`"#FF5733"`)

````python
x = np.arange(1,11)
y = x**2

plt.figure(figsize=(8,5))

plt.plot(x, y, color="green")

plt.title("Green Line")

plt.show()
````

## ➖ Changing Line Style

The `linestyle` parameter controls how the line is drawn.

Common styles:

- `"-"` → Solid
- `"--"` → Dashed
- `":"` → Dotted
- `"-."` → Dash-dot

````python
x = np.arange(1,11)
y = x**2

plt.figure(figsize=(8,5))

plt.plot(x, y, linestyle="--")

plt.title("Dashed Line")

plt.show()
````

## 🔵 Using Markers

Markers highlight individual data points.

Common markers:

- `"o"` → Circle
- `"s"` → Square
- `"^"` → Triangle
- `"*"` → Star
- `"x"` → Cross

````python
x = np.arange(1,11)
y = x**2

plt.figure(figsize=(8,5))

plt.plot(x, y, marker="o")

plt.title("Line with Circle Markers")

plt.show()
````

## 📏 Changing Line Width

The `linewidth` parameter controls the thickness of the plotted line.

A thicker line improves visibility in presentations and reports.

````python
x = np.arange(1,11)
y = x**2

plt.figure(figsize=(8,5))

plt.plot(x, y, linewidth=3)

plt.title("Thick Line")

plt.show()
````

## ✨ Combining Multiple Customizations

You can combine different customization options to create visually appealing plots.

````python
x = np.arange(1,11)
y = x**2

plt.figure(figsize=(8,5))

plt.plot(
    x,
    y,
    color="purple",
    linestyle="--",
    marker="o",
    linewidth=2
)

plt.title("Customized Line Plot")
plt.xlabel("Number")
plt.ylabel("Square")

plt.show()
````

## 🔲 Adding Grid

A grid makes it easier to read values from a graph.

Use the `plt.grid()` function to display horizontal and vertical grid lines.

````python
x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(8,5))

plt.plot(x, y)

plt.title("Line Plot with Grid")

plt.grid(True)

plt.show()
````

## 🏷️ Adding a Legend

A legend helps identify multiple datasets displayed on the same plot.

Use the `label` parameter inside `plt.plot()` and display it using `plt.legend()`.

````python
x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(8,5))

plt.plot(x, y, label="Square Numbers")

plt.title("Line Plot with Legend")

plt.legend()

plt.show()
````

## 📈 Plotting Multiple Lines

Matplotlib allows multiple datasets to be displayed on the same figure.

This is useful for comparing trends between different variables.

````python
x = np.arange(1, 11)

square = x ** 2
cube = x ** 3

plt.figure(figsize=(8,5))

plt.plot(x, square, label="Square")
plt.plot(x, cube, label="Cube")

plt.title("Square vs Cube")

plt.xlabel("Number")
plt.ylabel("Value")

plt.legend()

plt.grid(True)

plt.show()
````

## 🎨 Complete Example

Let's combine all the concepts learned so far to create a clean and professional visualization.

````python
x = np.arange(1,11)

y1 = x**2
y2 = x**3

plt.figure(figsize=(10,6))

plt.plot(
    x,
    y1,
    color="royalblue",
    linestyle="-",
    marker="o",
    linewidth=2,
    label="Square"
)

plt.plot(
    x,
    y2,
    color="tomato",
    linestyle="--",
    marker="s",
    linewidth=2,
    label="Cube"
)

plt.title("Square vs Cube Comparison", fontsize=16)

plt.xlabel("Number", fontsize=12)

plt.ylabel("Value", fontsize=12)

plt.grid(True)

plt.legend()

plt.show()
````
